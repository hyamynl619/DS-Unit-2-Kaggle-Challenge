````python
#Find duplicate columns 

def getDuplicateColumns(train):
  duplicateColumnNames = set()
  for x in range(train.shape[1]):
    col = train.iloc[:, x]
    for y in range(x + 1, train.shape[1]):
      otherCol = train.iloc[:, y]
      if col.equals(otherCol):
        duplicateColumnNames.add(train.columns.values[y])
        return list(duplicateColumnNames)
````

````python
# Get list of duplicate columns
duplicateColumnNames = getDuplicateColumns(train)
 
print('Duplicate Columns are as follows')
for col in duplicateColumnNames:
    print('Column name : ', col)
````

````python
#Wrangle Data

import numpy as np

def wrangle(X):
    
    X = X.copy()
    X['latitude'] = X['latitude'].replace(-2e-08, 0)
    
    cols_with_zeros = ['longitude', 'latitude']
    for col in cols_with_zeros:
        X[col] = X[col].replace(0, np.nan)
            
    X = X.drop(columns=['quantity_group', 'waterpoint_type'])
    
    return X


train = wrangle(train)
val = wrangle(val)
test = wrangle(test)

#I chose waterpoint_type also because it has a lot of similarities
````
